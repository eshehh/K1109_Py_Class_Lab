```python
#scatter
x1=[1,2,3,4,5]
x2=[10,12,14,16,18]

plt.scatter(x1,x2)

```

```python
x10=[x1,x2]
df1=pd.DataFrame(x10).T
df1.columns=['x1','x2']
df1
print(df1.x1.corr(df1.x2))
```

```python
x1=np.random.normal(1,1,size=(100,1))
x2=np.random.normal(-2,4,size=(100,1))
plt.scatter(x1,x2)
```

```python
x=np.concatenate((x1,x2), axis=1)
df=pd.DataFrame(x, columns=['x1','x2'])
print(df.x1.corr(df.x2))
```

```python
from datetime import datetime
filename='data/temperature_2014.csv'
df=pd.read_csv(filename)
df.PST=pd.to_datetime(df.PST)

fig=plt.figure(dpi=128, figsize=(10,6))
plt.plot(df.PST,df['Max TemperatureF'])
plt.plot(df.PST,df['Min TemperatureF'])
plt.fill_between(df.PST,df['Max TemperatureF'],df['Min TemperatureF'],facecolor='blue',alpha=0.2)
plt.title(" ",fontsize=20)
plt.xlabel('')
plt.ylabel('')
plt.tick_params(axis='both',labelsize=16)

plt.savefig('data/temperature_chart.png')
plt.show()
```
