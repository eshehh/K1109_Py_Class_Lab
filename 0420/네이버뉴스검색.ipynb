```python

nouns=komoran.nouns(aStr)
print(nouns)
```

```python
from collections import Counter
counter=Counter(nouns)
print(counter)
```

```python
counter2=counter.most_common(100)
print(counter2)
```

```python
wordInfo=dict()
for t,c in counter2:
    if (len(t)>1):
        wordInfo[t]=c
        # print(t,c)
dict(wordInfo).items()  
```

```python
taglist=pytagcloud.make_tags(dict(wordInfo).items(), maxsize=80)
taglist
```

```python
pytagcloud.create_tag_image(taglist, 'data/pytagcloud3.png',size=(800,800),fontname='Korean')
```
