# countplot()
- 항목의 갯수를 세어줌
- 알아서 column을 구성하고 있는 value들을 구분하여 보여줌

```python
sns.countplot(x='class', data=titanic,hue='who') # 가로로
# sns.countplot(y='class', data=titanic,hue='who') # 세로로
plt.show()
```

```python
# sns.set_palette('grey') # set_palette를 하면 그레이로 미리 지정
sns.countplot(x='class', data=titanic,hue='who', palette='spring_r')
```

https://www.geeksforgeeks.org/seaborn-color-palette/

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
sns.palplot(sns.color_palette("Greys"))
plt.show()
```

```python
sns.palplot(sns.color_palette("terrain_r", 7))
plt.show()
```

```python
# color=['Blue', 'White', 'Red', 'Yellow', 'Green', 'Grey']
sns.set_palette('deep')
sns.countplot(x='class', hue='who', data=titanic)
```

# 2. relplot
- 두 column간의 상관관계를 보여줌

```python
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='sex', size='size')
sns.set_palette('deep')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip',hue='smoker', row='sex', col='time', data=tips)
```

1. 온도문서를 가져와서 최고기온과 최저 기온 사이의 상관관계를 나타내는 그래프를 작성하라
2. taxis에서 total과 tip 사이의 상관관계 그래프를 작성하고, hue는 color를 지정하고 payment를
column으로 지정하는 subplot를 작성하라


```python
sns.get_dataset_names()
taxis=sns.load_dataset('taxis')
```

```python
sns.relplot(x='total', y='tip', data=taxis, hue='color',col='payment')
sns.set_palette('deep')
plt.show()
```
