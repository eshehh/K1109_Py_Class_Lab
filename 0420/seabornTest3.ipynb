```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
from matplotlib import rc,font_manager
font_path='C:/Windows/Fonts/gulim.ttc'
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)

plt.rcParams['figure.figsize']=(10,9)
plt.rcParams['axes.unicode_minus']=False #축의 마이너스 기호를 유니코드로 바꾸지마라
```

#Pairplot, Implot, jointplot 

```python
tips=sns.load_dataset('tips')
```

#Pairplot
- 그리드 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 표시
- 숫자형 columns에 대해서만 차트 작성

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='size')
plt.show()
```

# lmplot
- column 간의 선형관계를 확인하기에 용이한 차트

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', col='day', col_wrap=3)
```

# Jointplot
- scatter(산점도)와 histogram(분포)을 동시에 그려주며 숫자형 데이터만 표현가능

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```
