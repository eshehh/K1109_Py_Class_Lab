```python
df = pd.read_csv('고등어.csv', encoding='cp949')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Assuming the data is already stored in a DataFrame called 'df'
busan_data = df['부산']

# Create a line plot
plt.plot(busan_data)

# Calculate the trend line using polynomial regression
years = np.arange(len(busan_data))
coefficients = np.polyfit(years, busan_data, 1)
trend_line = np.polyval(coefficients, years)

# Plot the trend line
plt.plot(years, trend_line, color='red', linestyle='--')

# Set the labels and title
plt.xlabel('Years')
plt.ylabel('Number in Busan')
plt.title('Population in Busan over the years with Trend Line')

# Display the plot
plt.show()
```
