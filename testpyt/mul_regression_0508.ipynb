```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim


torch.manual_seed(1) 
```

```python
x1_train= torch.FloatTensor([[73], [93], [89], [96], [73]])
x2_train= torch.FloatTensor([[80], [88], [91], [98], [66]])
x3_train= torch.FloatTensor([[75], [93], [90], [100], [70]])
y_train= torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
# 가중치 w와 편향 b 초기화
w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
optimizer=optim.SGD([w1, w2, w3, b], lr=1e-5)
nb_epochs=100000

for epcho in range(nb_epochs):
    hypothesis=x1_train*w1 + x2_train*w2 + x3_train*w3 + b

    cost=torch.mean((hypothesis - y_train)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epcho%400==0:
        print('Epoch=', epcho, 'w1=', w1.item(), 'w2=', w2.item(), 'w3=', w3.item(), 'b=', b.item(), 'cost=', cost.item())
```

```python
x_train = torch.FloatTensor([[73, 80, 75],
[93, 88, 93],
[89, 91, 80],
[96, 98, 100],
[73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
W=torch.zeros((3,1), requires_grad=True)
b=torch.zeros(1, requires_grad=True)

optimizer=optim.SGD([W,b], lr=1e-5)

nb_epochs=2000

for epcho in range(nb_epochs):
    hypothesis=x_train.matmul(W)+b

    cost=torch.mean((hypothesis - y_train)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    if epcho%100==0:
        print('Epoch=', epcho, 'hypothesis=', hypothesis.squeeze().detach(), 'cost=', cost.item())
```

#nn.Modual 사용_single_regression

```python
torch.manual_seed(1)
# 데이터
x_train= torch.FloatTensor([[1], [2], [3]])
y_train= torch.FloatTensor([[2], [4], [6]])
```

```python
model=nn.Linear(1,1)
print(list(model.parameters()))

optimizer=torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
nb_epochs=2000

for epcho in range(nb_epochs):
    prediction=model(x_train)

    cost=F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epcho%100==0:
        print('epcho=', epcho, 'cost=', cost.item())
```

```python
new_data=torch.FloatTensor([[4.0]])

pred_y=model(new_data)
print("예측값:", pred_y)
```

# 단순선형회귀 클래스로 구현

```python
torch.manual_seed(1)
# 데이터
x_train= torch.FloatTensor([[1], [2], [3]])
y_train= torch.FloatTensor([[2], [4], [6]])
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(1,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model=LinearRegressionModel()
optimizer=torch.optim.SGD(model.parameters(), lr=0.01)

nb_epochs=2000
for epcho in range(nb_epochs):
    prediction=model(x_train)

    cost=F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epcho%100==0:
        print('epcho=', epcho, 'cost=', cost.item())
```

```python
x_train = torch.FloatTensor([[73, 80, 75],
[93, 88, 93],
[89, 91, 80],
[96, 98, 100],
[73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
class MulLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(3,1)

    def forward(self, x):
        return self.linear(x)    
```

```python
model=MulLinearRegressionModel()
optimizer=torch.optim.SGD(model.parameters(), lr=1e-5)

nb_epochs=100000

for epcho in range(nb_epochs):
    prediction=model(x_train)

    cost=F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epcho%500==0:
        print('epcho=', epcho, 'cost=', cost.item())
```
