```python
import numpy as np
```

```python
x_data1 = [[[73., 80., 75.],
            [93., 88., 93.],
            [89., 91., 80.],
            [96., 98., 100.],
            [73., 66., 70.]]]

t_data1 = [[152.],
           [185.],
           [180.],
           [196.],
           [142.]]
```

```python
x_data=np.array(x_data1)
t_data=np.array(t_data1)
print(x_data)
print(t_data)
```

```python
w=np.random.rand(3,1) #3행 1열짜리
b=np.random.rand(1)
print(w,b)
```

```python
#손실 함수 만들기
def loss_func(x, t):
    y=np.dot(x, w)+b # y=XW+b
    return (np.sum((t-y)**2))/len(x) #MSE(에러의 평균을 제곱한것의 합)
```

```python
def mul_numerical_derivative(f, x):
    delta_x=1e-4
    grad=np.zeros_like(x) # 미분결과 값 저장 배열

    it=np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        idx=it.multi_index

        tmp_val=x[idx]
        x[idx]=float(tmp_val)+delta_x
        fx1=f(x)
        x[idx]=tmp_val-delta_x
        fx2=f(x)
        grad[idx]=(fx1-fx2)/(2*delta_x)

        x[idx]=tmp_val
        it.iternext()
    return grad 
```

```python
def predict(x):
    y=np.dot(x,w)+b
    return y
```

```python
learning_rate=1e-6

f=lambda x:loss_func(x_data, t_data)
print("Initial error value=", loss_func(x_data, t_data), 'w=',w, 'b=',b)

for step in range(100000):
    w-=learning_rate*mul_numerical_derivative(f, w)
    b-=learning_rate*mul_numerical_derivative(f, b)
    if step%400==0:
        print("step", step, "error value", loss_func(x_data, t_data), "w=", w, 'b=',b)
```
